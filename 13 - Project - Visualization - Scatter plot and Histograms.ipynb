### Horsepower vs Torque
- Import matplotlib.pyplot as plt and use %matplotlib inline.
- Import csv
- Read the files/cars.csv wil into a list of dictionaries.
- Use list comprehension to get
    - Horsepwer
    - Torque
    - Bonus: Fuel City mpg
- Make a scatter plot
    - Options: Use alpha=.25 and c= with the list of Fuel City mpg

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import csv
```

```python
with open('files/cars.csv', 'r') as f: #abrimos el archivo en modo lectura y le asignamos la variable f
    csv_reader = csv.DictReader(f) #convertimos cada linea del archivo cxv en un diccionario
    cars = list(csv_reader)  #generamos una lista de diccionarios
```

para graficar hp vs torque, necesito dos listas, una lista con los valores de hp y otra con torque, tengo que sacara los datos de hp y torque de cada diccionario de la lista de diccionarios, y meterlos en una lista corresponeidnde

```python
hp = [int(data['Horsepower']) for data in cars]
torque = [int(data['Torque']) for data in cars]
fuel = [int(data['Fuel City mpg']) for data in cars]
```

```python
fig, ax = plt.subplots()
ax.scatter(hp, torque, c = fuel)
ax.set_title("HP vs Torque")
ax.set_xlabel("Horsepower")
ax.set_ylabel("Torque")
```

### Roll Dice
- Import random
- Create a functin role_dice(), which simulates a roll of 2 dice.
- Use list comprehension to create a list of 500 calls to role_dice().
- Create a histogram with 11 bins of the list.

```python
import random
```

```python
def role_dice():
    d1 = random.randint(1,6)
    d2 = random.randint(1,6)
    return d1 + d2
```

```python
datos = [role_dice() for i in range(0,500)]
```

```python
fig, ax = plt.subplots()
ax.hist(datos, bins=11)
ax.set_title("Role dice")
ax.set_xlabel("Resultado")
ax.set_ylabel("# repeticiones")
```
