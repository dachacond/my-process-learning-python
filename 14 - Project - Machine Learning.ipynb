### Description
- import numpy as np and import csv

- import matplotlib.pyplot as plt with %matplotlib inline
- Read the CSV file files/cars.csv as a list of dictionaries
- Use list comprehension to get Horsepower and Torque from the CSV data.

- Create a LinearRegression() model.
- Call **fit(.)** with the two reshaped lists.
- Call **predict(.)** with the reshaped list of Horsepower.
- Create a scatter plot with all the points (Horsepower vs Torque)
- Create a line (red) with the prediction result.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import csv
```

```python
with open('files/cars.csv', 'r') as f: #abrimos el archivo en modo lectura y le asignamos la variable f
    csv_reader = csv.DictReader(f) #convertimos cada linea del archivo cxv en un diccionario
    cars = list(csv_reader)  #generamos una lista de diccionarios
```

```python
hp = [int(data['Horsepower']) for data in cars]
torque = [int(data['Torque']) for data in cars]
fuel = [int(data['Fuel City mpg']) for data in cars]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Hp = np.array(hp).reshape((-1,1))
Torque = np.array(torque).reshape((-1,1))
```

```python
lin_regresor = LinearRegression()
lin_regresor.fit(Hp,Torque)

Y_pred = lin_regresor.predict(Hp)
```

```python
fig, ax = plt.subplots()
ax.scatter(hp, torque, c = fuel)
ax.plot(Hp, Y_pred, c ='r')
ax.set_title("HP vs Torque")
ax.set_xlabel("Horsepower")
ax.set_ylabel("Torque")
```
